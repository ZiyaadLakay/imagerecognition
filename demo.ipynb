### Importing libraries, models and demo data:

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import *
from sklearn.externals import joblib
```

```python
svmModel = joblib.load('svmModel.pkl') #SVM Model
logRegression = joblib.load('logisticModel.pkl') #Logistic Regression model
test = np.genfromtxt('group8_test.dat',delimiter=",",dtype="float32")

x_test = test[:,:12288]
y_test = test[:,-1]

Xrecons = x_test.reshape(240,64,64,3)
```

### Testing the DEMO :  
- It will ask the user to enter a number and then pick a specific image out of the data set or give the user the option of a random selection.
- The Program will read the image and predict which number is being held in the picture and display it to the user
- It will also show the Image after prediction
- It will then Ask the user if they would like to stop the program or continue to choose another number

```python
i = 1
j = 0



while(i != 0):
    print("")
    num = input("Choose Any Number in the Data Set (0-240) or press 'Enter' to generate a random number : ") 

    if(num == ""):
        num = randint(0,240)
        print("Random selection number "+str(num))
        predictLog = logRegression.predict([x_test[num]])
        predictSVM = svmModel.predict([x_test[num]])
    else:
        num = int(num)
        predictLog = logRegression.predict([x_test[num]])
        predictSVM = svmModel.predict([x_test[num]])

    print("The Original Label for the picture is : ",int(y_test[num]))
    print("The Predicted value Of the Picture Using Logistic Regression is : ",int(predictLog[0]))
    print("The Predicted value Of the Picture Using SVM is : ",int(predictSVM[0]))
    row = Xrecons[num]
    plt.imshow(row)
    plt.show()
    i = int(input("Would You like to Enter Another Number ? (0 - No ; 1 - Yes) : "))
print("Goodbye!")
```
